## Задание 1
Для каждого артиста вывести все жанры, которые есть в его песнях, и для каждого жанра вывести наиболее продолжительную песню этого артиста.

```python
%%sql
WITH Solution AS
(
  SELECT ar.name as Artist_name, g.name as Genre_name, t.name as Track_Name, t.Milliseconds, MAX(t.Milliseconds) OVER (PARTITION BY ar.ArtistID, g.GenreId) as MaxDurartion
  FROM artists ar
  JOIN albums al ON ar.ArtistId = al.ArtistId
  JOIN tracks t ON al.AlbumId = t.AlbumId
  JOIN genres g ON t.GenreId = g.GenreId
)
SELECT Artist_name, Genre_name, Track_Name, Milliseconds
FROM Solution
WHERE Milliseconds = MaxDurartion
ORDER BY Artist_name
```

## Задание 2
Посчитать скользящее среднее выручки по годам.


```python
%%sql
WITH RevenueByYear AS (
    SELECT strftime('%Y', invoices.InvoiceDate) AS Year, SUM(tracks.UnitPrice * invoice_items.Quantity) AS YearlyRevenue
    FROM invoices
    JOIN invoice_items ON invoices.InvoiceId = invoice_items.InvoiceId
    JOIN tracks ON invoice_items.TrackId = tracks.TrackId
    GROUP BY strftime('%Y', invoices.InvoiceDate)
)
SELECT Year, YearlyRevenue,
    AVG(YearlyRevenue) OVER (
        ORDER BY Year ASC
        ROWS BETWEEN 1 PRECEDING AND CURRENT ROW
    ) AS Average
FROM RevenueByYear;
```

## Задание 3

Для каждого плейлиста вывести трек с наибольшей продолжительностью, трек с наименьшей стоимостью (если у треков одинаковая стоимость, то вывести тот, который весит меньше всех), и жанр данного плейлиста (жанром плейлиста будет называться тот жанр, треков которого больше всех в плейлисте).

```python
%%sql
WITH PlaylistTracks AS (
    SELECT pl.PlaylistId,pl.Name AS PlaylistName,t.TrackId,t.Name AS TrackName,t.Milliseconds,t.UnitPrice,t.Bytes,g.GenreId,g.Name AS GenreName,
        COUNT(*) OVER (PARTITION BY pl.PlaylistId, g.GenreId) AS GenreTrackCount,
        RANK() OVER (PARTITION BY pl.PlaylistId ORDER BY t.Milliseconds DESC) AS DurationRank,
        ROW_NUMBER() OVER (PARTITION BY pl.PlaylistId ORDER BY t.UnitPrice ASC, t.Bytes ASC) AS CostRank
    FROM playlists pl
    JOIN playlist_track plt ON pl.PlaylistId = plt.PlaylistId
    JOIN tracks t ON plt.TrackId = t.TrackId
    JOIN genres g ON t.GenreId = g.GenreId
)
SELECT PlaylistId, PlaylistName,
    MAX(CASE WHEN DurationRank = 1 THEN TrackName END) AS MaxDurationTrackName,
    MAX(CASE WHEN DurationRank = 1 THEN Milliseconds END) AS MaxDuration,
    MAX(CASE WHEN CostRank = 1 THEN TrackName END) AS MinCostTrackName,
    MAX(CASE WHEN CostRank = 1 THEN UnitPrice END) AS MinCost,
    MAX(CASE WHEN CostRank = 1 THEN Bytes END) AS MinCostTrackSize,
    MAX(CASE WHEN GenreTrackCount = (
        SELECT MAX(GenreTrackCount)
        FROM PlaylistTracks pt2
        WHERE pt2.PlaylistId = PlaylistTracks.PlaylistId
    ) THEN GenreName END) AS MostPopularGenre
FROM PlaylistTracks
GROUP BY PlaylistId, PlaylistName;
```

## Задание 4

Для каждого жанра вывести треки, которые встречались в большем числе заказов,
сумму, потраченную на данные треки, количество купленных треков и компанию, которая больше всех приобрела данную композицию.

```python
%%sql

WITH TrackOrders AS (
    SELECT
        g.Name AS GenreName,
        t.Name AS TrackName,
        SUM(ii.Quantity) AS TotalPurchases,
        SUM(ii.Quantity * t.UnitPrice) AS TotalRevenue,
        c.Company AS CompanyName
    FROM
        invoice_items AS ii
    JOIN tracks AS t ON ii.TrackId = t.TrackId
    JOIN genres AS g ON t.GenreId = g.GenreId
    JOIN invoices AS i ON ii.InvoiceId = i.InvoiceId
    JOIN customers AS c ON i.CustomerId = c.CustomerId
    GROUP BY
        g.Name, t.Name, c.Company
),
RankedTracks AS (
    SELECT
        GenreName,
        TrackName,
        TotalPurchases,
        TotalRevenue,
        CompanyName,
        ROW_NUMBER() OVER (PARTITION BY GenreName ORDER BY TotalPurchases DESC, TotalRevenue DESC) AS rnk
    FROM TrackOrders
)
SELECT
    GenreName,
    TrackName,
    TotalPurchases,
    TotalRevenue,
    CompanyName
FROM RankedTracks
WHERE rnk = 1 and CompanyName IS NOT NULL
ORDER BY GenreName;
```

## Задание 5
Составить рейтинг клиентов по количеству потраченных ими денег.

```python
%%sql
WITH CustomerSpending AS (
  SELECT c.CustomerId, c.FirstName, c.LastName, SUM(ii.Quantity * t.UnitPrice) AS TotalSpent
  FROM customers c
  JOIN invoices i ON c.CustomerId = i.CustomerId
  JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
  JOIN tracks t ON ii.TrackId = t.TrackId
  GROUP BY c.CustomerId
)
SELECT RANK() OVER (ORDER BY TotalSpent DESC) as P, FirstName, LastName, TotalSpent
FROM CustomerSpending
ORDER BY TotalSpent DESC;
```
